# MT copy number 

```python
import os
import csv
import sys

path = 'offtarget'

fileOut=open('offtarget.WXS.csv',"w")
fileOut.write('ID,cov\n')

for f in os.listdir(path):
    file=open("./offtarget/"+f)
    reader=csv.reader(file)
    cov=0.0
    count=0
    for line in reader:
            x=int(line[1])
            y=int(line[2])
            if y>x:
                if line[3]=="":
                    n=0.0
                else:
                    n=float(line[3])
                cov+=n/(y-x)
                count+=1
        
    fileOut.write(f.replace('.offtarget.cov','')+","+str(cov/count))
    fileOut.write("\n")
```

```python
import pandas as pd
metadata=pd.read_csv('../outcomes/PILOT.WXS.clean.metadata.csv')
df_MT = pd.read_csv('MT.WXS.csv')

df_MT.head()

df = pd.read_csv('offtarget.WXS.csv')
data=pd.merge(df_MT, metadata, on='ID')
data_cov_WXS=pd.merge(df, data, on='ID')


data_cov_WXS['cov_MT_norm'] = data_cov_WXS['cov.MT']/data_cov_WXS['cov']
data_cov_WXS.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.boxplot(x="disease", y="cov_MT_norm", data=data_cov_WXS)
g=sns.stripplot(x="disease", y="cov_MT_norm", data=data_cov_WXS, palette=["black"], jitter=0.05)
g.set(xlabel='Cancer type', ylabel='Normalized MT dosage')
g=sns.despine()
#plt.show(g)
plt.savefig("MT.cov.pdf", bbox_inches='tight')
```

```python
import pandas as pd
metadata=pd.read_csv('../outcomes/PILOT.RNASeq.clean.metadata.csv')
df_MT = pd.read_csv('MT.RNASeq.csv')
data_MT=pd.merge(df_MT, metadata, on='ID')

depth_RNASeq=pd.read_csv('../bams/NR.RNASeq.csv')
data_RNASeq_cov=pd.merge(data_MT, depth_RNASeq, on='ID')

data_RNASeq_cov['cov.MT.RNASeq.norm'] = (data_RNASeq_cov['cov.MT.RNASeq']/data_RNASeq_cov['NR'])*1000000


data_MT.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.boxplot(x="disease", y="cov.MT.RNASeq.norm", data=data_RNASeq_cov,order=["BRCA", "LUAD", "LUSC","PRAD","SKCM"])
g=sns.stripplot(x="disease", y="cov.MT.RNASeq.norm", data=data_RNASeq_cov, palette=["black"], jitter=0.05,order=["BRCA", "LUAD", "LUSC","PRAD","SKCM"])
g.set(xlabel='Cancer type', ylabel='Normalized MT dosage (RNA-Seq)')
g=sns.despine()
#plt.show(g)
plt.savefig("MT.cov.RNASeq.pdf", bbox_inches='tight')
```

```python
data_RNASeq_cov2 = data_RNASeq_cov[['patientBarcode','cov.MT.RNASeq','cov.MT.RNASeq.norm']]
data_cov_WXS2=data_cov_WXS[['patientBarcode','cov_MT_norm','disease']]
data_cov_both=pd.merge(data_RNASeq_cov2, data_cov_WXS2, on='patientBarcode')
data_cov_both.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.regplot(x="cov_MT_norm", y="cov.MT.RNASeq.norm", data=data_cov_both,color="black")
g.set(xlabel='cov_MT_norm', ylabel='cov.MT.RNASeq.norm')
g=sns.despine()
#plt.show(g)
plt.savefig("MT.cov.WXS.vs.RNASeq.pdf", bbox_inches='tight')

stats.pearsonr(data_cov_both['cov_MT_norm'], data_cov_both['cov.MT.RNASeq.norm'])
```
